## Prime numbers

```python
def isPrime(num1):

    #check for Integer value if not it returns TypeError.
    try:
        if num1<=1:
            print('Non prime')
        elif num1==2:
            print('Prime')
        else:
            #condition to check prime number.
            for i in range(2,num1):
                if num1%i==0:
                    print('Non prime')
                    return
            print('Prime')

    except TypeError:
        print(f'{num1} is not an Integer')
    return
    
```

```python
isPrime(4)
```

## Product of Random Numbers

```python
import numpy as np
```

## Squares of Even/Odd Numbers

```python
m1=[]                       #intialize empty list

for i in range(100,201):
    if i%2==0:              #check for even or not
        m1.append(i**2)     #adding square of each number to the list m1   

print(f'Square of even numbers from 100 to 201 is {m1}')
```

## Word counter

```python
from collections import Counter

def wordCounter(text=""):
    m1=text.split()            #split words from the text
    return Counter(m1)         #Count the repeated words
```

```python

text = "This is a sample text. This text will be used to demonstrate the word counter."
wordCounter(text)
```

## Check for Palindrome

```python
x="porararop"

def isPallindrome(word):
    
    word=str(word).lower().strip()         #converts word into smaller case and remove white spaces.
    reverse=""                             #initilize empty string to store reverse of x
    
    for i in range(len(word)-1,-1,-1):
        reverse=reverse+word[i]            #reversing the string and concate in same string 'reverse'
    
    if word==reverse:                      #checking condition
        print('Palindrome')
    else:
        print('Not palindrome')
    
```

```python
p="23RTrtr32"
isPallindrome(p)
```
