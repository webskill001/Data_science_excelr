```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#data points given
df = pd.Series([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
mean = df.mean() #calculating sample mean
mean
```

```python
std = df.std()   #calculating sample standard deviation
std
```

```python
"""
constructing a 99% confidence interval for the mean number of characters printed 
before the print-head fails using the sample standard deviation
"""
conf_interval = stats.t.interval(0.99, 14,
                                 loc = mean,
                                 scale = std)
print(f'Build 99% Confidence Interval Using Sample Standard Deviation is {np.round(conf_interval, 3)}')
```

```python
"""
constructing a 99% confidence interval for the mean number of characters printed 
before the print-head failure using the population standard deviation 0.2
"""
conf_interval_Known_std = stats.t.interval(0.99, 14,
                                 loc = mean,
                                 scale = 0.2)
print(f'99% Confidence Interval Using Known Population Standard Deviation is {np.round(conf_interval_Known_std, 3)}')
```
