## Visualization

### Histograms

```python
plt.figure(figsize=(15,8), facecolor='lightblue')

plt.subplot(3,2,1)
plt.hist(df['Volume'], bins=6)
plt.xlabel('Volume')


plt.subplot(3,2,2)
plt.hist(df['Avg Price'], bins=6)
plt.xlabel('Avg Price')

plt.subplot(3,2,3)
plt.hist(df['Total Sales Value'], bins=8)
plt.xlabel('Total Sales Value')

plt.subplot(3,2,4)
plt.hist(df['Discount Rate (%)'], bins=8)
plt.xlabel('Discount Rate (%)')

plt.subplot(3,2,5)
plt.hist(df['Discount Amount'], bins=5)
plt.xlabel('Discount Amount')

plt.subplot(3,2,6)
plt.hist(df['Net Sales Value'], bins=7)
plt.xlabel('Net Sales Value')

plt.show()
```
