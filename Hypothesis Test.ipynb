```python
#import the libraries
import pandas as pd
import scipy 
import numpy as np
from scipy import stats
```

# Super Market example

```python
p_value = (130-120)/(40/np.sqrt(80))
p_value
```

```python
1-stats.t.cdf(p_value,df=79)   #cdf always gives us left side of area 
```

# Call Center example

```python
p_value = (4.6-4)/(3/np.sqrt(50))
p_value
```

```python
2*stats.t.cdf(-p_value,df=49)   # '-' sign added becoz cdf find left side area 
```

# One tail test

```python
data=pd.Series([0.593, 0.142, 0.329, 0.691, 0.231, 0.793, 0.519, 0.392, 0.418])
```

```python
#'two sided','less','greater'
```

```python
scipy.__version__
```

```python
#This gives two tail test p value so divide by 2 to get final p-value
p=stats.ttest_1samp(data,0.3, alternative='greater')
p
```

# Two Tail test

```python
Control=pd.Series([91, 87, 99, 77, 88, 91])
Treat =pd.Series([101, 110, 103, 93, 99, 104])  
```

```python
stats.ttest_ind( Control,Treat, alternative='two-sided')
```

# 2 Proportion Test

```python
import numpy as np
```

```python
#Data:

n1 = 1000
p1 = .4

n2 = 1000
p2 = .39
```

```python
population1 = np.random.binomial(1, p1, n1)
population2 = np.random.binomial(1, p2, n2)
```

```python
population1
```

```python
import statsmodels.api as sm
sm.stats.ttest_ind(population1, population2)    #as we increase sample pvalue is very very low even if probability diff is 0.005
```
