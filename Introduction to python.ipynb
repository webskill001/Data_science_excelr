```python



def even_odd_frac(l):
    m1=[]
    m2=[]
    m3=[]
    for i in l:
        if i%2==0:
            m1.append(i)
        elif i%2==1:
            m2.append(i)
        else:
            m3.append(i)

    return m1,m2,m3


even_odd_frac([1,2,3,4,5,6,3.4])
            
```

```python
def isPrime(num):

    if num<=1:
        return False

    if num==2:
        return True
        
    for i in range (2,num):
        if num%i==0:
            return False
    
    
    return True
```

```python
isPrime(2)
```

## list comprehension

```python
l=[1,2,3,4,5,6]
m=[]

for i in l:
    m.append(i*2)

m
```

```python
i=[1,2,3,4,5,6]
m=[2*i for i in l]   #list comprehension
m
```

```python
l=[1,8,9,10,45]
m=['even' if i%2==0 else 'odd' for i in l]
m
```

```python
l=[1,8,9,10,45]
m=[i for i in l if i%2==0]
m
```

## map function

```python
l=[1,8,9,10,5.6]

def isEven(num):
    
    if num%2==0:
        return 'even'
    elif num%2==1:
        return 'odd'

isEven(5)
```

```python
m=map(isEven,l)
list(m)
```

## Exception handling

```python
def division(x,y):
    
    try:
        return x/y
    except:
        print('error')
    finally:
        print('finally')
```

```python
division(34,0)
```
