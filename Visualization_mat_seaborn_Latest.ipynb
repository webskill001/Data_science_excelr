# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips
```

```python
tips['size'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.show()
```

## Univariate  one variable data visualization

```python
#strip plot

sns.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='size', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.boxplot(x='size', y='tip', data=tips, palette=['red','yellow','orange','lightblue','green','blue'])
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(data=tips, x='time', y='tip',jitter=True)
plt.show()
```

```python
sns.boxplot(data=tips,x='time', y='tip')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip',hue='sex', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python
plt.figure(figsize=(20,10),facecolor='lightblue')
x=np.linspace(-4,4,50)



'''
this create me the space where 5 rows and 1 column
subplot(numOfRows, numOfColumns, indexofGraph)

'''
plt.subplot(5,1,1)
plt.plot(x,x**2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot 1', size=20)

plt.subplot(5,1,2)
plt.plot(x,np.tan(x))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot 2 tan', size=20)

plt.subplot(5,1,3)
plt.plot(x,np.sin(x))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot 3 sin', size=20)

plt.subplot(5,1,4)
plt.plot(x,np.cos(x))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot 4 cos', size=20)

plt.subplot(5,1,5)
plt.plot(x,np.tanh(x))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot 4 tanh', size=20)

plt.show()
```

```python
survived = data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived");
```

```python
sns.pairplot(data.dropna());
```

```python
# Pclass vs Survive
grid = sns.FacetGrid(data, col='Survived', row='Pclass', size=2.4, aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python

sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()
plt.title("QueensTown")
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
plt.title("Southampton")
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
#Most of the Passengers embarked from Southampton
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
```
